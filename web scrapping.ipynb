## Création d'une fonction qui récupère toutes les informations sur une offre

```python
def get_info(link):
    page = requests.get(link)
    soup = bs(page.text, "lxml")
    company = soup.find('h3', attrs = {'class' : "sc-12bzhsi-11 jhkNVT"}).string
    title = soup.find('h1', attrs = {'class': "sc-12bzhsi-3 kuNRyl"}).string
    lieu = soup.find('span', attrs = {'class': "sc-16yjgsd-3 cToOtz"}).string 
    contrat = soup.find('span', attrs = {'class': "sc-16yjgsd-3 bCCdzk"}).span.string
    spans = soup.find_all('span', class_="sc-16yjgsd-3 eJxlVj")
    education = spans[0].find_next("span").find_next("span").string
    experience = spans[1].find_next("span").find_next("span").string
    try : 
        if soup.find("time").has_attr('datetime'):
            debut= soup.find("time").string
    except : 
        debut = "Non trouvé"
            
    col = soup.find_all('ul', attrs = {'class': "sc-16yjgsd-4 gbiZfI"})
    for c in col : 
        domaine = c.find_next('span', class_="sc-16yjgsd-3 eJxlVj").string
        taille =  c.find_next('span', class_="sc-16yjgsd-3 cToOtz").string
    description = soup.find("div", class_="itvpid-1 erXZUZ").text

    line=[company,domaine,taille,lieu,title,debut,contrat,education,experience,link]
    return line 

a = get_info("https://www.welcometothejungle.com/fr/companies/datascientest/jobs/designer-graphique-illustration-h-f-cdi_puteaux_DATAS_2ZaOmVr?q=299b35fef71fd5cb48e04ec81f61587a&o=947840")
a
```
